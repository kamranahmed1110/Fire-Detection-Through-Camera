```python
#putting every thing in the function
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()
    # extract filename
    image_name = root.find('filename').text
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')

    # Check if 'objs' list is not empty
    if objs:
        # can store every information in the parser
        parser = []
        obj = objs[0]
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])
    else:
        # Handle the case where no 'object' elements are found
        parser = None

    return parser


```

```python
#center x, center y
df['center_x'] = ((df['xmax']+df['xmin'])/2)/df['width']
df['center_y'] = ((df['ymax']+df['ymin'])/2)/df['height']
```

```python
df['xmax'] = pd.to_numeric(df['xmax'], errors='coerce')
df['xmin'] = pd.to_numeric(df['xmin'], errors='coerce')
df['width'] = pd.to_numeric(df['width'], errors='coerce')
df['ymax'] = pd.to_numeric(df['ymax'], errors='coerce')
df['ymin'] = pd.to_numeric(df['ymin'], errors='coerce')
df['height'] = pd.to_numeric(df['height'], errors='coerce')

df['center_x'] = ((df['xmax'] + df['xmin']) / 2) / df['width']
df['center_y'] = ((df['ymax'] + df['ymin']) / 2) / df['height']
df['w'] = (df['xmax'] - df['xmin']) / df['width']
df['h'] = (df['ymax'] - df['ymin']) / df['height']
```

```python
#saving images and label in text
import os
from shutil import move
```

```python
train_folder = 'Datacluster Fire and Smoke Sample/train'
test_folder = 'Datacluster Fire and Smoke Sample/test'

os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
cols = ['filename', 'id' , 'center_x' , 'center_y','w','h']
groupby_obj_train = train_df[cols].groupby('filename')
groupby_obj_test = test_df[cols].groupby('filename')
```

```python
#groupby_obj_train.get_group('Datacluster Fire and Smoke Sample (1).jpg').to_csv('sample.txt',index=False,header=False)
#save each image in train/test folder and respective label in .txt file
def save_data(filename, folder_path, group_obj):
    #move image
    src = os.path.join('Datacluster Fire and Smoke Sample',filename)
    dst = os.path.join(folder_path,filename)
    move(src,dst) #move image to the destination folder
    #save label information
    text_filename = os.path.join(folder_path,
                                 os.path.splittext(filename)[0]+'.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename,sep=' ',index=False,header=False)
```
